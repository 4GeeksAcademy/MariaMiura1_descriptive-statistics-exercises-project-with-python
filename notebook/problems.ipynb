### Exercise 1

We will use NumPy to obtain information to describe statistically.

- Generate an array of 100 elements following a normal distribution.
- Generate an array of 100 elements following a chi-square distribution with 3 degrees of freedom.
- Calculate the main metrics and statistical measures that best describe the two vectors.

````python
import statistics as stats

from scipy.stats import skew
from scipy.stats import kurtosis

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt


normal_distribution = np.random.normal(size=100)
chi = np.random.chisquare(3, 100)

print("Normal distribution", str(normal_distribution), "Chi", str(chi))
#Media
print(f"Normal distribution: {stats.mean(normal_distribution)}")
print(f"Chi mean: {stats.mean(chi)}")
#Mediana
print(f"Normal median: {stats.median(normal_distribution)}")
print(f"Chi median: {stats.median(chi)}")


#Standar deviation
print(f"Starndar deviation normal: {np.var(normal_distribution)}Standar deviation chi: {np.var(chi)}")
#min max
#
````

### Exercise 2

Write a Python program to calculate the standard deviation of the following data:

```py
data = [4, 2, 5, 8, 6]
```

````python
import matplotlib.pyplot as plt
from matplotlib import rcParams


def sd_calculation(data):
    n = len(data)
    if n <=1: return 0, 0
    mean = sum(data) / n    
    variance = sum((x - mean) ** 2 for x in data) / (n - 1)
    return variance ** 0.5, mean
data = [4, 2, 5, 8, 6]
std_deviation, mean = sd_calculation(data)

print(f"Standar deviation: {std_deviation}") 

plt.plot(data, label='Data', marker='o', color='cyan', linestyle='-', markersize=8)  # Datos en cian con círculos
plt.axhline(mean, color='magenta', linestyle='--', linewidth=2, label=f'Mean ({mean})')  # Línea horizontal de la media en magenta


plt.gca().set_facecolor('lightpink')  
plt.grid(True, linestyle='-.', color='green', alpha=0.5)  
plt.xlabel('♡Index♡', fontsize=12, color='purple')
plt.ylabel('♡Value♡', fontsize=12, color='orange')
plt.title('♡Data Plot with the Mean♡', fontsize=14, color='darkviolet')

plt.text(2, 7, '❤️', fontsize=20, color='gold', ha='center')  
plt.text(3, 8, '❤️', fontsize=20, color='red', ha='center') 
plt.text(1, 3, '❤️', fontsize=20, color='purple', ha='center')  
plt.text(4, 6, '❤️', fontsize=20, color='blue', ha='center') 

plt.xlim(-0.5, len(data)-0.5)
plt.ylim(min(data)-1, max(data)+1)

plt.legend(facecolor='lightblue', loc='best')


plt.show()
````
